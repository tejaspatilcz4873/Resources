# What is a Series?
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
#  Creating a pandas series from a list 

a = [1, 7, 2, 34, 5432, 433,'Texas', 'Ohio']
series1 = pd.Series(a)
print(series1)
print()

print(series1[3]) #the labels(indexes) mentioned can be used to access a specified value.

print()
print(type(series1)) #checking the datatype for the variable series1
```

```python
# With the index argument, you can name your own labels/indexex.
# if labels/indexes aren't specifically defined, by default labels/indexes would start with zero asimilar to lists.



b = [1, 7, 2, 34, 5432, 433, 'Texas', 'Ohio', 'Dakota', 'New Mexico']
series1 = pd.Series(b, index = [1, 3, 2, 5, 25, 15, 67, 45, 41, 87])
print(series1)

print()
print(series1[67])

# Note : In case the list/dictionary you want to create series from, the datatype of highest priority is assigned as datatype of the series.
# For example, presence of float and int, float gets priority. Similary, for presence of float, int, str, series is assigned str as datatype.
```

```python
# You can also use a key/value object, like a dictionary, when creating a Series.
# The keys of the dictionary become the labels.

import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)

print()
print(myvar['day3'])
```

```python
#To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.

import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

myvar
#print(myvar)
```

# DataFrames
1. A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns
2. Series is like a column, a DataFrame is the whole table.
3. Effectively, dataframe is like a collection of multiple series.
4. Dataframes can be created from csv files, excel files, json files, python dictionary files as well.

```python
# Creating a dataframe

# Way 1 - passing a nested list as a data argument. The lists present inside may or may not be of same length.
# If lengths are different, NaN value is assigned by default

df1 = pd.DataFrame(data = [[1,3,4454.45,21], [32,'Nevada',52,678,67,345]])
df1
```

```python
# Transposing the dataframe

df2 = df1.transpose()
print(df2)

print()

df2
df1

# Note the contrast in the way data is shown on screen with/without the use of print method 
```

```python
# Checking the default column names

print(df2.columns) # default column indexing/naming
```

```python
# Assigning/modifying the column names for convenience

df2

df2.columns = ['column_x', 'column_y']

df2
```

```python
# Checking the column names after modification

print(df2.columns) # default column naming

print(type(df2))
```

```python
# Filtering the data from dataframe

df[df['Fare'] > 50] # Got a temp dataframe that has all rows from original dataframe with fare greater than 50

# As result suggests, there are 160 passengers that paid fare greater than 50.
# This result can be again stored permanently in a new dataframe variable.
```

```python
# Filtering data for passengers with age greater than 40 and paid more than 60 dollars as fare'
# Remember to use the bitwise operators for joining multiple conditions instead of logical operators(Use of '&' instead of 'and').

'''
Operator	Name	
& 	         AND	
|	         OR	
 ^	         XOR	
~ 	         NOT	
'''

df_new = df[(df['Age'] > 40) & (df['Fare'] > 60)]
df_new
```

```python
df_new.shape
```

```python
# grouping data in a dataframe
'''
Pandas groupby is used for grouping the data according to the categories 
and applying a function to the categories. It also helps to aggregate data efficiently.
'''

# getting count of male and female passengers.
group1 = df.groupby('Sex') # created a group
group1['PassengerId'].count() #calculated the count of passengerIds for the categories based on group.
```

```python
# grouping data by multiple columns

group2 = df.groupby(['Sex', 'SibSp']) # created a group
group2['PassengerId'].count() #calculated the count of passengerIds for the categories based on group.
```

```python
# For each gender, find total passenger count and the total fare collected.
'''
The aggregate() method allows you to apply a function or a list of function names to be executed along one of the axis of the DataFrame,
default 0, which is the index (row) axis.

Note: the agg() method is an alias of the aggregate() method.
'''

group3 = df.groupby(['Sex']) #created the group
print(group3.agg({'PassengerId':'count', 'Fare':'sum'})) #under agg() function, passed all the aggegates that I wanted to calculate


# Using above method, you won't be able to define the column headers/names for readability for the output dataframe. 
# Hence, we can also use below method where we can also define column headers for readability.
print()
print(group3.agg(Passenger_Count = ('PassengerId', 'count'), Total_Fare = ('Fare', 'sum')))
```

```python
# Operation performed in cell 53 above, same operation can also be performed using agg() method as depicted below
# Also the output of these operations would also be a dataframe over which we can query/perform further operations

group2 = df.groupby(['Sex', 'SibSp']) # created a group
df3 = group2.agg(Passenger_Count = ('PassengerId','count')) #calculated the count of passengerIds for the categories based on group.

print(df3)
```
