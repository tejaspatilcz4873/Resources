# Matplotlib: Visualization with Python

```python
# Importing the library
# Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.

import matplotlib.pyplot as plt
```

# Plotting x and y points

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
# Draw a line in a diagram from position (1, 3) to position (8, 10).

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# To plot only the markers instead of line, you can use shortcut string notation parameter 'o', which means 'rings'.


plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
# You can plot as many points as you like, just make sure you have the same number of points in both axis.

# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10).

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Marking just the points and not the complete line.
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

# Default X-Points

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
# So, if we take the same example as above, and leave out the x-points, the diagram will look like this.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

# Markers
You can use the keyword argument marker to emphasize each point with a specified marker. Many markers like star, circle, point, pixel, 'X', etc are available for use.

```python
# Mark each point with a circle

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Mark each point with a star

plt.plot(ypoints, marker = '*')
plt.show()
```

# Format Strings fmt

You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|c lineolorolor

```python
# Mark each point with a circle, join points with dotted line with color red

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, '*:r') 
# '*' is marker, ':' represents dotted line, 'r' represent the red color

plt.show()
```

# Marker Size

You can use the keyword argument markersize or the shorter version, ms to set the size of the markers.

```python
# You can use the keyword argument markersize or the shorter version, ms to set the size of the markers.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '^', ms = 20)
plt.show()
```

# Marker-edge Color

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
# Set the EDGE color to red.

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

# Markerface color

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*', ms = 10, mfc = 'r')
plt.show()
```

```python
# Use both the mec and mfc arguments to color the entire marker:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r') # bot mec and mfc provided 'r' as argument for solid fill.
plt.show()
```

# You can also use Hexadecimal color values:

You can use hexadecimal codes for the colors as an argument to fmt, mec and mfc

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

```python
# Or any of the 140 supported color names.

# Mark each point with the color named "hotpink":

plt.plot(ypoints, marker = '^', ms = 20, mec = 'hotpink', mfc = 'hotpink')
```

# Linestyle

You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line.

```python
plt.plot(ypoints, marker = '^', ms = 20, mec = 'orange', mfc = 'silver', ls = 'dotted')
```

```python
plt.plot(ypoints, marker = '^', ms = 20, mec = 'indigo', mfc = 'silver', ls = 'dashed')
```

# Line Color

You can use the keyword argument color or the shorter c to set the color of the line:

```python
plt.plot(ypoints, marker = '^', ms = 20, mec = 'indigo', mfc = 'silver', ls = 'dashdot', color = 'maroon')
```

# Line Width

You can use the keyword argument linewidth or the shorter lw to change the width of the line.

The value is a floating number, in points.

```python
plt.plot(ypoints, marker = '^', ms = 20, mec = 'indigo', mfc = 'silver', ls = 'dashdot', linewidth = '5.0', color = 'magenta')
```

# Multiple Lines

You can plot as many lines as you like by simply adding more plt.plot() functions:

```python
# Draw two lines by specifying a plt.plot() function for each line:

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10, 14, 16])
y2 = np.array([6, 2, 7, 11, 13, 9])

plt.plot(y1)
plt.plot(y2)

#plt.show()
```

```python
# Draw two lines by specifiyng the x- and y-point values for both lines:

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
# You can also insert the labels for both these lines into the legend for better reference
# Note that you won't be able apply labels for multiple lines in a single plot() instance for multiple lines in previous code-block.

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10, 14, 16])
y2 = np.array([6, 2, 7, 11, 13, 9])

plt.plot(y1, label = 'line_plot_1') # adding the label for legend
plt.plot(y2, label = 'line_plot_2')

plt.legend() # showing the legend

plt.show()
```

# Matplotlib Labels and Title

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

With Pyplot, you can use the title() function to set a title for the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Average Pulse vs Calorie Burnage")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

# Add Grid Lines to a Plot

With Pyplot, you can use the grid() function to add grid lines to the plot.

You can use the axis parameter in the grid() function to specify which grid lines to display.

Legal values are: 'x', 'y', and 'both'. Default value is 'both'.

```python

import matplotlib.pyplot as plt

x = [80, 85, 90, 95, 100, 105, 110, 115, 120, 125]
y = [240, 250, 260, 270, 280, 290, 300, 310, 320, 330]

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(axis = 'x')
plt.show()
```

# Set Line Properties for the Grid

You can also set the line properties of the grid, like this: 

grid(color = 'color', linestyle = 'linestyle', linewidth = number).

```python
import matplotlib.pyplot as plt

x = [80, 85, 90, 95, 100, 105, 110, 115, 120, 125]
y = [240, 250, 260, 270, 280, 290, 300, 310, 320, 330]

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(axis = 'both', color = 'magenta', linestyle = 'dotted', linewidth = 2.5)
plt.show()
```

# Matplotlib Subplotlot.

```python
# Draw 2 plots:

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) # the figure has 1 row, 2 columns, and this plot is the first plot. 
# In simple words the display area would be divided into one row and two columns and those sections the subplots would fit in.
plt.plot(x,y)
plt.title('Graph 1') # adding title for subplot 1
plt.xlabel('x-label_1')
plt.ylabel('y-label_1')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)
plt.title('Graph 2') # adding title for subplot 2
plt.xlabel('x-label_2')
plt.ylabel('y-label_2')


plt.suptitle('Figure Super-Title') # Adding supertitle for the whole figure


plt.show()
```

```python
# Draw 2 plots on top of each other:

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) # we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# Draw 3 plots on top of each other:

import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 1, 1) # we want a figure with 3 rows an 1 column (meaning that the three plots will be displayed on top of each other instead of side-by-side)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 1, 2) # 2(third argument) here represent that this would be the second plot of the subplot sequence.
plt.plot(x,y)


#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([60, 70, 80, 90])

plt.subplot(3, 1, 3)
plt.plot(x,y)


plt.show()
```

```python
# Draw 6 plots

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 2, 1)
plt.plot(x,y, marker = '*')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 2, 2)
plt.plot(x,y, marker = '*')

x = np.array([5, 6, 7, 8])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 2, 3)
plt.plot(x,y, marker = '*')

x = np.array([9, 10, 11, 12])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 2, 4)
plt.plot(x,y, marker = '*')

x = np.array([13, 14, 15, 16])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 2, 5)
plt.plot(x,y, marker = '*')

x = np.array([17, 18, 19, 20])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 2, 6)
plt.plot(x,y, marker = '*')

plt.show()
```
