# Visualize Distributions With Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# Relational plots in Seaborn

Relational plots are used for visualizing the statistical relationship between the data points. Visualization is necessary because it allows the human to see trends and patterns in the data. The process of understanding how the variables in the dataset relate each other and their relationships are termed as Statistical analysis. 

To draw the relational plots seaborn provides three functions. These are:
1. relplot()
2. scatterplot()
3. lineplot()


Both scatterplots and lineplots can also be drawn with the help of kind parameter in relplot(). Basically relplot(), by default, gives us scatterplot() only, and if we pass the parameter kind = “line”, it gives us lineplot().

```python
# Importing dataset using pandas

import pandas as pd

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

df1
```

```python
# Relplot example

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (6,3))

sns.relplot(x = df1['Age'], y = df1['Fare'])

plt.show()

# Used matplotlib.pyplot here to adjust the figure size. Not mandatory to use as such
```

# KDE Plot Visualization with Pandas and Seaborn

Kernel Density Estimate (KDE) plot, a visualization technique that offers a detailed view of the probability density of continuous variables.

It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization. It provides a smoothed representation of the underlying distribution of a dataset.

The KDE plot visually represents the distribution of data, providing insights into its shape, central tendency, and spread. It is particularly useful when dealing with continuous data or when you want to explore the distribution without making assumptions about a specific parametric form (e.g., assuming the data follows a normal distribution).

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

sns.kdeplot(x = df1['Fare'], hue = df1['Sex'])
```

# Seaborn Histplot

Seaborn Histplot is used to visualize the univariate set of distributions(single variable). It plots a histogram, with some other variations like kdeplot and rugplot.

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

sns.histplot(x = df1['Fare'], hue = df1['Pclass'], bins = 25)

# since y parameter wasn't passed, it used the value count as default y parameter. Same phenomenon happens for other plots as well in Seaborn.

# Passed an additional paramter bins to adjust the intervals for plotting. 
# Due to this bins parameter, we're currently observing 14 bars in the graph.
```

```python
# Passing kde parameter value as True display a kernel density estimate (KDE) of the data in addition to the histogram

import pandas as pd
import seaborn as sns

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

sns.histplot(x = df1['Age'], hue = df1['Sex'], kde = True, bins = 14)

# Passed an additional paramter bins to adjust the intervals for plotting. 
# Due to this bins parameter, we're currently observing 14 bars in the graph.
```

# seaborn.pairplot()

To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function. 

The diagonal plots are the univariate plo(variable_interval vs count)ts, and this displays the relationship for the (n, 2) combination of variables in a DataFrame as a matrix of plots.

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

sns.pairplot(data = df1)
```

# seaborn.barplot() method

A barplot is basically used to aggregate the categorical data according to some methods and by default it’s the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv(r"C:\Users\Tejas Patil\Desktop\IT Learning\Growthschool\Python\Titanic\titanic.csv")
df1 = df.dropna().head(100)

sns.barplot(x = df1['Sex'], y = df1['Fare'], hue = df1['Pclass'])
```
